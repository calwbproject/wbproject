```python
import pandas as pd
import sqlite3


conn = sqlite3.connect('../db.sqlite3')
cur = conn.cursor()

cur.execute("select * from whiteboard_engineer")

rows = cur.fetchall()
```

```python
df = pd.DataFrame(rows)

filterd_by_name = df[~df[1].str.contains(r'[\u4E00-\u9FFF]', regex=True)]

filterd_by_name[8] = 'グローバル'

filterd_by_name
```

```python
for idx, row in filterd_by_name.iterrows():
    engineer_id = row[0]
    engineer_type = row[8]
    cur.execute(
        "update whiteboard_engineer set engineer_type = ? where id = ?",
        (engineer_type, engineer_id)
    )
    
conn.commit()
conn.close()

```
