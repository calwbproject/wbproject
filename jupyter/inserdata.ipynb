```python
import sqlite3

conn = sqlite3.connect('../db.sqlite3')

cur = conn.cursor()
```

```python
cur.execute('select * from auth_user;')

rows = cur.fetchall()
#id	password	'last_login'	'is_superuser'	'username'	'last_name'	'email'	'is_staff'	'is_active'	'date_joined'
columns = [desc[0] for desc in cur.description]

df_user = pd.DataFrame(rows, columns=columns)
df_user
```

```python
dropduple_df = df['sales'].drop_duplicates()
dropduple_df = dropduple_df.reset_index(drop=True)
dropduple_df
```

```python
df_user['first_name']
```

```python
firstname =[row['first_name'] for idx, row in df_user.iterrows()]
for i in range(len(dropduple_df)):
    df_user.loc[i+1, 'first_name'] = dropduple_df[i]
    
df_user
```

```python
for idx, row in df_user.iterrows():
    cur.execute(
        "update auth_user set first_name = ? where id = ?",
        (row['first_name'], row['id'])
    )
    
conn.commit()
```
